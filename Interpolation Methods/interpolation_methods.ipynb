Nearest neighbour Interpolation

```python
import numpy as np
from urllib.request import urlopen
import cv2
import matplotlib.pyplot as plt
from PIL import Image

'''image = np.asarray(bytearray(urlopen("https://cdn.lesnumeriques.com/optim/focus/article/20/20205/comment-augmenter-la-taille-d-une-photo-22f912e3__312_176__overflow.jpg").read()), dtype="uint8")
image = cv2.imdecode(image, cv2.IMREAD_COLOR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)'''

image = cv2.imread("color_image.JPG")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
resize_shape = (image.shape[1] * 2, image.shape[0] * 2)

image_ = cv2.resize(image, resize_shape, interpolation=cv2.INTER_NEAREST)

plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original Image")
plt.subplot(1,2,2)
plt.imshow(image_)
plt.title("Nearest Neighbour Interpolated")

```

Bilinear Interpolation

```python
image_ = cv2.resize(image, resize_shape, interpolation=cv2.INTER_LINEAR)
plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original Image")
plt.subplot(1,2,2)
plt.imshow(image_)
plt.title("Bilinnear Interpolated")
```

Bicubic Interpolation

```python
image_ = cv2.resize(image, resize_shape, interpolation=cv2.INTER_CUBIC)
plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original Image")
plt.subplot(1,2,2)
plt.imshow(image_)
plt.title("Bicubic Interpolated")
```

Lanczos Interpolation

```python
image_ = cv2.resize(image, resize_shape, interpolation=cv2.INTER_LANCZOS4)
plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original Image")
plt.subplot(1,2,2)
plt.imshow(image_)
plt.title("lanczos Interpolated")
```
