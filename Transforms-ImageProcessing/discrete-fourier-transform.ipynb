```python
import numpy as np
import matplotlib.pyplot as plt
```

plot a sine wave

```python
x = np.arange(-100, 101)
wavelength = 50
amp = np.sin(2 * np.pi * x/ wavelength)
plt.plot(x, amp)
```

Convert 1D sine wave to a 2D grating

```python
X, Y = np.meshgrid(x, x)
grating = np.sin(2 * np.pi * X/ wavelength)
plt.imshow(grating, cmap='gray')
```

2d grating with different orientation

```python
angle =np.pi/4 #(45 degree)
grating_rotated = np.sin(2*np.pi*(X * np.cos(angle) + Y * np.sin(angle))/ wavelength)
plt.imshow(grating_rotated, cmap='gray')
```

Fourier Transform of the gratings

```python
ft = np.fft.ifftshift(grating)  # shift the grating and fft to the centre for better visualization. If not shifted, it remains in the top left corner 
ft = np.fft.fft2(grating)
ft = np.fft.fftshift(ft)
plt.subplot(1,2,1)
plt.imshow(grating, cmap='gray')
plt.title("grating")
plt.subplot(1,2,2)
plt.imshow(abs(ft), cmap='gray')
plt.title("fourier Transform")
plt.xlim(90, 110) #zoom at the center for better visualization
plt.ylim(110, 90) 
plt.show()


ft = np.fft.ifftshift(grating_rotated)  
ft = np.fft.fft2(grating_rotated)
ft = np.fft.fftshift(ft)
plt.subplot(2,2,1)
plt.imshow(grating_rotated, cmap='gray')
plt.title("rotated grating")
plt.subplot(2,2,2)
plt.imshow(abs(ft), cmap='gray')
plt.title("fourier Transform")
plt.xlim(90, 110)
plt.ylim(110, 90) 
plt.rcParams['figure.figsize'] = [60, 40]
plt.show()
```

Mix two gratings

```python
mix_grating = grating + grating_rotated
ft = np.fft.ifftshift(mix_grating)  
ft = np.fft.fft2(mix_grating)
ft = np.fft.fftshift(ft)
plt.subplot(2,2,1)
plt.imshow(mix_grating, cmap='gray')
plt.title("mixed grating")
plt.subplot(2,2,2)
plt.imshow(abs(ft), cmap='gray')
plt.title("fourier Transform")
plt.xlim(90, 110)
plt.ylim(110, 90) 
plt.rcParams['figure.figsize'] = [60, 40]
plt.show()
```

Fourier Transform of an Image

```python
from urllib.request import urlopen
from PIL import Image, ImageOps
image = Image.open(urlopen("https://images.unsplash.com/photo-1592339078319-1f8cc9d67888?ixlib=rb-4.0.3&ixid=MnwxMjA3fDB8MHxzZWFyY2h8Mnx8ZGFpc3klMjBmbG93ZXJ8ZW58MHx8MHx8&w=1000&q=80"))
image = image.resize([200, 200])
image = ImageOps.grayscale(image=image)
plt.imshow(image)

ft = np.fft.ifftshift(image)  
ft = np.fft.fft2(image)
ft = np.fft.fftshift(ft)
plt.subplot(2,2,1)
plt.imshow(image, cmap='gray')
plt.title("image")
plt.subplot(2,2,2)
plt.imshow(abs(ft), cmap='gray')
plt.title("fourier Transform")
plt.xlim(90, 110)
plt.ylim(110, 90)
plt.rcParams['figure.figsize'] = [60, 40]
plt.show()
```
