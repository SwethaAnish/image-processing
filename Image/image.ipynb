Import Required Packages and Create a numpy array with all values 0. 

```python
import numpy as np
from PIL import Image
import cv2
import matplotlib.pyplot as plt
image = np.empty([100, 100], dtype=np.uint8)
image.fill(0)
image_ =  Image.fromarray(image)
image_
```

Create a white image using numpy

```python
image.fill(250)
image_ = Image.fromarray(image)
image_
```

Create a checkerboard Image

```python
check = True
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        if check:
            image[i ,j] = 0
        else:
            image[i, j] = 255
        if j % 20 == 0:
            check = not(check)
    if i % 20 != 0:
        check = not(check)
image_ = Image.fromarray(image)
image_
```

Convert a Graysclae image to Binary Image

```python
inp_image  = cv2.imread("../Basic Image Processing Techniques/bg_image.png", 0)
inp_image = cv2.cvtColor(inp_image, cv2.COLOR_RGB2BGR)
ret, binary_image = cv2.threshold(inp_image, 150, 255, cv2.THRESH_BINARY)
binary_image = Image.fromarray(binary_image)
plt.subplot(1,2,1)
plt.imshow(inp_image)
plt.subplot(1,2,2)
plt.imshow(binary_image)
plt.show
```

Convert a RGB Image to Grayscale Image

```python
inp_image  = cv2.imread("../Basic Image Processing Techniques/bg_image.png")
inp_image = cv2.cvtColor(inp_image, cv2.COLOR_BGR2RGB)
gray_img =  cv2.cvtColor(inp_image, cv2.COLOR_BGR2GRAY)
plt.subplot(2, 1, 1)
plt.imshow(inp_image)
plt.subplot(2, 1, 2)
plt.imshow(gray_img, cmap='gray')
plt.show
print(gray_img)
```

Visualize 3 channels of an Image 

```python
image = cv2.imread("../Image/color_image.JPG")
#convert the image to RGB (images are read in BGR in Opencv2)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#split the image into its three channels
(R,G,B) = cv2.split(image)
plt.subplot(1,3,1)
plt.imshow(R, cmap='gray')
plt.title("RED CHANNEL")
plt.subplot(1,3,2)
plt.imshow(G, cmap='gray')
plt.title("GREEN CHANNEL")
plt.subplot(1,3,3)
plt.imshow(B, cmap='gray')
plt.title("BLUE CHANNEL")
plt.rcParams['figure.figsize'] = [120, 80]
plt.show()
```
